```python
some_energies_consumption = ["nuclear_consumption","oil_consumption","fossil_fuel_consumption","solar_consumption","hydro_consumption","wind_consumption","gas_consumption"]
consumption_df=df[some_energies_consumption]
consumption_df.loc[(df['country'] == 'Canada') & (df['year'] == 2018)]
```

```python
turkey_2018 = consumption_df.loc[(df['country'] == 'Turkey') & (df['year'] == 2018)]
total_energy_consumption = turkey_2018['nuclear_consumption'] + turkey_2018['oil_consumption'] + turkey_2018['solar_consumption']+turkey_2018["fossil_fuel_consumption"]+turkey_2018["hydro_consumption"]+turkey_2018["wind_consumption"]+turkey_2018["gas_consumption"]
print("Türkiye'nin 2018'deki toplam enerji tüketimi:", total_energy_consumption.sum())
x=round(total_energy_consumption.sum(),2)
print("yuvarlanmış hali",x)
```

```python
df_2019 = df[df['year'] == 2019]
df_2018 = df[df['year'] == 2018]
toplam_enerji_tuketimi_2018 = df_2018[some_energies_consumption].sum()
toplam_enerji_tuketimi_2019 = df_2019[some_energies_consumption].sum()
```

```python
toplam_enerji_tuketimi_2019
```

```python
plt.figure(figsize=(6,4))
toplam_enerji_tuketimi_2019.plot(kind='bar', color='skyblue')
plt.title('total energy consumption')
plt.xlabel('types of energy')
plt.ylabel('total consumption')
plt.xticks(rotation=45)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].pie(toplam_enerji_tuketimi_2018, labels=toplam_enerji_tuketimi_2018.index, autopct='%1.1f%%', startangle=90)
axs[0].set_title('Energy consumption in 2018')

axs[1].pie(toplam_enerji_tuketimi_2019, labels=toplam_enerji_tuketimi_2019.index, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Energy consumption in 2019')

plt.show()
```

```python
import geopandas as gpd

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
for index, row in world.iterrows():
    country_name = row['name']
    country_2018 = consumption_df.loc[(df['country'] == country_name) & (df['year'] == 2019)]
    
    total_energy_consumption = country_2018['nuclear_consumption'] + country_2018['oil_consumption'] + country_2018['solar_consumption'] + country_2018["fossil_fuel_consumption"] + country_2018["hydro_consumption"] + country_2018["wind_consumption"] + country_2018["gas_consumption"]
    
    total = total_energy_consumption.sum()
    rounded_total = round(total, 2)
    
    print(f"{country_name}'nin 2018'deki toplam enerji tüketimi:", rounded_total)
```

```python
enerji_tuketimi_2019 = {}  

for index, row in world.iterrows():
    country_name = row['name']
    country_2019 = consumption_df.loc[(df['country'] == country_name) & (df['year'] == 2019)]
    
    total_energy_consumption = country_2019['nuclear_consumption'] + country_2019['oil_consumption'] + country_2019['solar_consumption'] + country_2019["fossil_fuel_consumption"] + country_2019["hydro_consumption"] + country_2019["wind_consumption"] + country_2019["gas_consumption"]
    
    total = total_energy_consumption.sum()
    enerji_tuketimi_2019[country_name] = round(total, 2)

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

world['Total Energy Consumption 2019'] = world['name'].map(enerji_tuketimi_2019)

fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(column='Total Energy Consumption 2019', ax=ax, legend=True,
           legend_kwds={'label': "2019 energy consumption", 'orientation': "horizontal"})
plt.title('Energy Consumption of World Countries in 2019')
plt.show()
# Since the country names in the data set and the country names in the geopandas library are not the same, the data of some countries appears as 0.
```
