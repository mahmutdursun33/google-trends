```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('trends.csv')
df.head()
```

```python
df.shape
```

```python
df.dropna()
```

```python
konum= 'Turkey'
yil= 2020
trend = df[(df['location'] == konum) & (df['year'] == yil)]['query'].value_counts().head(5)
print(yil,konum," 5 trend arama sonucu" ":")
print(trend)
```

```python
kelime = 'Aşure'
sorgu= df[df['query'] == kelime].groupby('year')['rank'].sum()
sorgu.plot(kind='line', marker='o', xlabel='yil', ylabel='hacim', title='' + kelime)
plt.show()
# dünyada aşure kelimesinin arama hacmi
```

```python
konums= 'Turkey'
trends = df[(df['location'] == konums) & (df['category'] =='Dizi')]['query'].value_counts().head(5)
print(konums," 5 trend arama sonucu" ":")
print(trends)
```

```python
features = ['location', 'category', 'query']
sel_year = 2020

sel_region = 'Turkey'
df_filter = df[(df.location==sel_region) & (df.year==sel_year)]

for f in features:
    plt.figure(figsize=(10,4))
    df_filter[f].value_counts()[0:25].plot(kind='bar')
    plt.title('Year = ' + str(2020) + ' / ' + f)
    plt.grid()
    plt.show()
```
